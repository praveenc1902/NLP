```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
print(STOP_WORDS)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("We just opened our wings, the flying part is coming soom")
```

```python
for token in doc:
  if token.is_stop:
    print(token)
```

```python
def preprocess(text):
  doc = nlp(text)
  no_stop_words = [token.text for token in doc if not token.is_stop and not token.is_punct ]
  return ' '.join(no_stop_words)
```

```python
print(preprocess("We just opened our wingss, flying part is coming soom buddy"))
```

```python
print(preprocess("The other is no other but divine brother"))
```

```python
import pandas as pd
df = pd.read_json("doj_press.json",lines=True)
df.head()
```
